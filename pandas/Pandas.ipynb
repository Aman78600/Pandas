# Groupby Functions

```python
cricket={
    'team':['india','india','australia','sa','sa','sa','sa','nz','nz','nz','nz','india'],
    'rank':[2,3,1,2,3,4,1,1,2,4,1,2],
    'year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
    'pooints':[876,801,891,815,776,784,834,824,758,691,883,782]
}
df=pd.DataFrame(cricket)
df
```

```python
df.groupby('team').groups
```

```python
df.groupby(['team','year']).get_group(('india',2017))
```

```python
df.groupby('team').sum()['pooints']
```

```python
df.groupby('team').sum()['pooints'].sort_values()
```

```python
df.groupby('team').sum()['pooints'].sort_values(ascending=False)
```

```python
g=df.groupby('team')
g['pooints'].agg([np.sum,np.mean,np.std,np.max,np.min])
```

```python
df
```

```python
df.groupby('team').filter(lambda x: len(x)==4)
```
